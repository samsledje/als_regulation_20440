## UpSet

```python
from upsetplot import plot, from_contents
```

```python
edge_contents = {
   "iPSC_ALS": list(ipsc_als_net.edges()),
   "iPSC_CTR": list(ipsc_ctr_net.edges()),
   "iMN_ALS": list(imn_als_net.edges()),
   "iMN_CTR": list(imn_ctr_net.edges()),
}
```

```python
sns.set(style='whitegrid',font_scale=1.5)
plot(from_contents(edge_contents),sort_categories_by=None)
plt.savefig(f"{FIG_DIR}/four_network_upset.pdf",bbox_inches='tight')
plt.show()
```
